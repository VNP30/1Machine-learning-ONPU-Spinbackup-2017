```python
import graphlab
```

```python
data = graphlab.SFrame('Mass Shootings Dataset.csv')
```

```python
data.head()
```

```python
def transform_mental(l):
    if l == 'Yes':
        return 0
    elif l == 'No':
        return 1
    elif l == 'Unclear':
        return 2
    elif l == 'Unknown':
        return 2
    else:
        return l
```

```python
data['Mental Health Issues']=data['Mental Health Issues'].apply(transform_mental)
```

```python
data
```

    На основе примера обучить модель с использованием такого набора фич:

[‘Total victims’, ’Mental Health Issues’] для

    3 кластеров

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(data['Total victims'], '.r')
```

```python
model = graphlab.kmeans.create(data, features=['Total victims', 'Mental Health Issues'],
                               num_clusters=3)
```

```python
model['cluster_id'].head()
```

```python
graphlab.canvas.set_target('ipynb')
model.show()
```

```python
data['cluster'] = model['cluster_id']['cluster_id']
data
```

```python
# Clustered data
plt.plot(data[data['cluster'] == 0]['Total victims'], '^r')
plt.plot(data[data['cluster'] == 1]['Total victims'], 'sb')
plt.plot(data[data['cluster'] == 2]['Total victims'], 'og')

# Centers
plt.plot(model['cluster_info']['Total victims'], 'hy')
```
